```python
import random
import json
import requests
import pandas as pd
import configparser
import matplotlib.pyplot as plt
import gmaps
from pprint import pprint
config = configparser.ConfigParser()
#config.read('../../API_keys/config.py')
config.read('../../../API_keys/config.py')
gkey = config['DEFAULT']['G_KEY']
gmaps.configure(api_key=gkey)
```

```python
cities_df = pd.read_csv('../Resources/Cities_dataframe.csv')
cities_df.dropna(inplace = True) 
cities_df.head()
```

```python
locations = cities_df[["Lat", "Lng"]]
humidity = cities_df["Humidity"].astype(float)
```

```python
fig = gmaps.figure()
heat_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=10, point_radius=1)
fig.add_layer(heat_layer)
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# sort by 
  * A max temperature lower than 80 degrees but higher than 70.

  * Wind speed less than 10 mph.

  * Zero cloudiness.
```

```python
ideal_cities_df = cities_df.loc[cities_df["Max Temp"] < 80]
ideal_cities_df = ideal_cities_df.loc[ideal_cities_df["Max Temp"] > 70]
ideal_cities_df = ideal_cities_df.loc[ideal_cities_df["Cloudiness"] == 0]
ideal_cities_df
```
