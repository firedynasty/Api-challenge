### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
ideal_cities_df_countries = ideal_cities_df["Country"]
ideal_cities_df_countries
```

```python
#create a dataframe
hotel_df = ideal_cities_df[["City","Lat","Lng","Country"]]
hotel_df
```

```python
hotel_df['Hotel Name'] = ""
hotel_df['Address'] = ""
```

```python
base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
params = {
    "location": "0,0",  # philadelphia coords,
    "type": "lodging",
    "radius": 5000,
    "key": gkey,
}

# use iterrows to iterate through pandas dataframe
for index, row in hotel_df.iterrows():
   
    first = hotel_df.iloc[index,1]
    second = hotel_df.iloc[index,2]
    
    target_coordinates = f"{first},{second}"
    
    # add keyword to params dict
    params['location'] = target_coordinates

    # assemble url and make API request
    print(f"Retrieving Results for Index {index}: {row[0]}.")
    response = requests.get(base_url, params=params).json()
    
    # extract results
    results = response['results']
    
    try:
        print(f"Closest hotel to {row[0]} is {results[0]['name']}.")
        hotel_df.loc[index, 'hotel name'] = results[0]['name']
        hotel_df.loc[index, 'address'] = results[0]['vicinity']
        
    except (KeyError, IndexError):
        print("Missing field/result... skipping.")
        
    print("------------")
```

```python
hotel_df = hotel_df.rename(columns={"hotel name": "Hotel Name", "address": "Address"})
```

```python
#create dataframe to insert the hotel information

# NOTE: Do not change any of the code in this cell

# Using the template add the hotel marks to the heatmap
info_box_template = """
<dl>
<dt>Name</dt><dd>{Hotel Name}</dd>
<dt>City</dt><dd>{City}</dd>
<dt>Country</dt><dd>{Country}</dd>
</dl>
"""
# Store the DataFrame Row
# NOTE: be sure to update with your DataFrame name
hotel_info1 = [info_box_template.format(**row) for index, row in hotel_df.iterrows()]
locations1 = hotel_df[["Lat", "Lng"]]
```

```python
#create layer
marker_layer = gmaps.marker_layer(locations1, info_box_content=hotel_info1)
fig.add_layer(marker_layer)
fig
```
